**Download all txt data files from gva website**

Careful, it takes several hours to download over 1700 files and 600MB of data

```python
from pathlib import Path
import time
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

# Compose absolute path to "Data" folder in your current working directory
data_folder = str(Path.cwd() / "Data")

service = Service(executable_path='geckodriver.exe')
options = Options()
options.set_preference("browser.download.folderList", 2)
options.set_preference("browser.download.dir", data_folder)
options.set_preference("browser.helperApps.neverAsk.saveToDisk", "text/plain,application/octet-stream")
options.set_preference("pdfjs.disabled", True)

driver = webdriver.Firefox(service=service, options=options)

driver.get('https://webcat-web.gva.es/webcat_web/datosHistoricosRvvcca/cargarDatosHistRvvcca#')
time.sleep(5)

for year in range(1994, 2026):
    print(f"Processing year: {year}")
    select_year = Select(driver.find_element(By.ID, "anyo"))
    select_year.select_by_visible_text(str(year))
    time.sleep(1)  # Wait for dropdown to update if needed

    download_button = driver.find_element(By.ID, "cargarDatosTabla")
    download_button.click()
    time.sleep(2)  # Wait for the page to load after clicking the button

    descargar_links = driver.find_elements(By.XPATH, '//a[@class="descargar"]')
    print(f"Number of 'descargar' links found for {year}: {len(descargar_links)}")

    for link in descargar_links:
        tipo = link.get_attribute('tipo_fich')
        if tipo == "mh":
            filename = link.get_attribute('nom_fich_hist')
            print(f"Downloading: {filename} (tipo_fich={tipo})")
            driver.execute_script("arguments[0].scrollIntoView();", link)
            time.sleep(0.5)
            link.click()
            time.sleep(1)  # Wait a bit between downloads if needed

    # Optionally, wait for downloads to finish for this year
    time.sleep(2)

# Wait for all downloads to finish
time.sleep(10)
driver.quit()
```
